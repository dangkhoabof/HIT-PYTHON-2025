ðŸ§© BÃ i 2 â€“ Xá»­ lÃ½ chuá»—i (String)

YÃªu cáº§u:
Viáº¿t chÆ°Æ¡ng trÃ¬nh nháº­n vÃ o má»™t chuá»—i, sau Ä‘Ã³:

Äáº¿m sá»‘ lÆ°á»£ng nguyÃªn Ã¢m vÃ  phá»¥ Ã¢m trong chuá»—i.

In ra chuá»—i Ä‘áº£o ngÆ°á»£c.

Kiá»ƒm tra xem chuá»—i cÃ³ pháº£i lÃ  palindrome (Ä‘á»c xuÃ´i ngÆ°á»£c nhÆ° nhau) khÃ´ng.

```python
nguyenAm="euoai"
chuoi=input("nhap 1 chuoi: ")
sokytuNguyenAm=0
soKyTuPhuAm=0
for kyTu in chuoi:
    if kyTu.isalpha():
        if kyTu in nguyenAm:
            sokytuNguyenAm+=1
        else:
            soKyTuPhuAm+=1
chuoiDao=chuoi[::-1]
print("so luong nguyen am: ",sokytuNguyenAm)
print("so luong phu am: ",soKyTuPhuAm)
print("chuoi dao nguoc la: ",chuoiDao)
if chuoiDao.strip()==chuoi.strip():
    print("chuoi la chuoi palindrome")
else:
    print("chuoi khong la chuoi palindrome")
```

ðŸ§© BÃ i 3 â€“ Tá»•ng há»£p kiáº¿n thá»©c

YÃªu cáº§u:
Viáº¿t chÆ°Æ¡ng trÃ¬nh nháº­p má»™t cÃ¢u tiáº¿ng Viá»‡t, sau Ä‘Ã³:

TÃ¡ch cÃ¢u thÃ nh cÃ¡c tá»« (dá»±a vÃ o dáº¥u cÃ¡ch).

Chuyá»ƒn toÃ n bá»™ vá» chá»¯ thÆ°á»ng.

Loáº¡i bá» cÃ¡c tá»« trÃ¹ng láº·p (chá»‰ giá»¯ má»™t láº§n).

Lá»c ra cÃ¡c tá»« cÃ³ Ä‘á»™ dÃ i â‰¥ 4 kÃ½ tá»±.

In ra danh sÃ¡ch tá»« duy nháº¥t, sáº¯p xáº¿p theo báº£ng chá»¯ cÃ¡i, vÃ  sá»‘ lÆ°á»£ng tá»« há»£p lá»‡.

```python
chuoi=input("nhap mot chuoi: ")
tu=[d for d in chuoi.split()]
tu=[t.lower() for t in tu]
tu=[d for d in tu if len(d)>=4]
tu_khong_lap=set(tu)
tu_khong_lap=list(tu_khong_lap)
tu_khong_lap.sort()
tu_khong_lap=" ".join(tu_khong_lap)
print(tu_khong_lap)
```
