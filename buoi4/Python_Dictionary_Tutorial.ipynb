## üîß 2. T·∫°o dictionary

`````python

# C√°ch 1: D√πng d·∫•u ngo·∫∑c nh·ªçn {}
person = {"name": "Bob", "age": 25}

# C√°ch 2: D√πng h√†m dict()
person2 = dict(name="John", age=30)
person2.update(name="Jane", age=35)

# C√°ch 3: T·∫°o dict r·ªóng r·ªìi th√™m ph·∫ßn t·ª≠
data = {}
person2["name"] = "Hanoi"
person2["age"] = 15

print(person)
print(person2)
print(data)
`````

## üéØ 3. Truy c·∫≠p v√† s·ª≠a gi√° tr·ªã

`````python
# dictionary kh√¥ng ƒë∆∞·ª£c truy xu·∫•t qua ch·ªâ s·ªë (index) m√† th√¥ng qua kh√≥a (key).
# Truy xu·∫•t b·∫±ng [key]: a[2], a["name"], ...
# Truy xu·∫•t b·∫±ng get(key): a.get(2), a.get("name")
# Th√™m/S·ª≠a m·ª•c: <t√™n_dict>[<key>] = <value>

person = {"name": "Alice", "age": 20, "city": "Hanoi"}

# Truy c·∫≠p
print(person["name"])        # Alice
print(person.get("age"))     # 20
print(person.get("phone", "Kh√¥ng c√≥"))  # Tr√°nh l·ªói n·∫øu key kh√¥ng t·ªìn t·∫°i

# C·∫≠p nh·∫≠t
person["age"] = 21
person["email"] = "alice@gmail.com"
print(person)
`````

## üß© 4. X√≥a ph·∫ßn t·ª≠

`````python

student = {"name": "Nam", "grade": "A", "year": 2025}

# pop(key) - x√≥a v√† tr·∫£ v·ªÅ gi√° tr·ªã
student.pop("grade")

# popitem() - x√≥a c·∫∑p cu·ªëi c√πng (Python 3.7+)
student.popitem()

# del - x√≥a theo key
del student["name"]

# clear() - x√≥a to√†n b·ªô
student.clear()

print(student)
`````

## üîÅ 5. Duy·ªát qua dictionary

`````python

scores = {"Alice": 85, "Bob": 90, "Charlie": 78}

# Duy·ªát key
for name in scores:
    print(name)

# Duy·ªát value
for score in scores.values():
    print(score)

# Duy·ªát c·∫£ key v√† value
for name, score in scores.items():
    print(name, ":", score)

`````

## üß∞ 6. C√°c ph∆∞∆°ng th·ª©c h·ªØu √≠ch

`````python

info = {"name": "Lan", "age": 19}

# copy()
clone = info.copy()

# update() - g·ªôp dict kh√°c
info.update({"city": "Hue"})

# keys(), values(), items()
print(info.keys()) # tr√≠ch xu·∫•t v√† tr·∫£ v·ªÅ c√°c kh√≥a c·ªßa c√°c m·ª•c trong t·ª´ ƒëi·ªÉn
print(info.values()) # tr√≠ch xu·∫•t v√† tr·∫£ v·ªÅ c√°c gi√° tr·ªã c·ªßa c√°c m·ª•c trong t·ª´ ƒëi·ªÉn
print(info.items()) # tr√≠ch xu·∫•t v√† tr·∫£ v·ªÅ c√°c c·∫∑p (key, value) c·ªßa c√°c m·ª•c trong t·ª´ ƒëi·ªÉn

# fromkeys() - t·∫°o dict t·ª´ danh s√°ch key
# dict.fromkeys(alphabets,number): t·∫°o m·ªôt t·ª´ ƒëi·ªÉn t·ª´ chu·ªói kh√≥a v√† gi√° tr·ªã  ƒë√£ cho
# alphabets ‚Äì d·ªØ li·ªáu c√°c kh√≥a ƒë∆∞·ª£c l∆∞u tr·ªØ b·ªüi chu·ªói, t·∫≠p h·ª£p, danh s√°ch,...
# numbers (t√πy ch·ªçn) ‚Äì l√† gi√° tr·ªã thu·ªôc b·∫•t k·ª≥ ki·ªÉu d·ªØ li·ªáu n√†o.
# Gi√° tr·ªã gi·ªëng nhau ƒë∆∞·ª£c g√°n cho t·∫•t c·∫£ c√°c kh√≥a c·ªßa t·ª´ ƒëi·ªÉn
# M·∫∑c ƒë·ªãnh gi√° tr·ªã None ƒë∆∞·ª£c g√°n cho c√°c kh√≥a c·ªßa t·ª´ ƒëi·ªÉn n·∫øu kh√¥ng c√≥ gi√°  tr·ªã ƒë∆∞·ª£c truy·ªÅn v√†o.

keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)
print(new_dict)

`````

## üí° 7. Dictionary comprehension

`````python

# T·∫°o dict t·ª´ comprehension
# { <bi·ªÉu_th·ª©c_key> : <bi·ªÉu_th·ª©c_value>  for <bi·∫øn> in <iterable>  if <ƒëi·ªÅu_ki·ªán_t√πy_ch·ªçn> }

squares = {x: x**2 for x in range(5)}
print(squares)

# L·ªçc ph·∫ßn t·ª≠
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)

`````

## üßÆ 8. B√†i t·∫≠p th·ª±c h√†nh

Trong Python, h√†m `sorted()` d√πng ƒë·ªÉ **s·∫Øp x·∫øp d·ªØ li·ªáu** v√† **tr·∫£ v·ªÅ m·ªôt danh s√°ch m·ªõi ƒë√£ s·∫Øp x·∫øp**, kh√¥ng l√†m thay ƒë·ªïi d·ªØ li·ªáu ban ƒë·∫ßu.

#### **C√∫ ph√°p c∆° b·∫£n**

```python
sorted(iterable, key=None, reverse=False)
````

| Tham s·ªë    | √ù nghƒ©a                                                  |
| ---------- | -------------------------------------------------------- |
| `iterable` | D·ªØ li·ªáu c√≥ th·ªÉ duy·ªát ƒë∆∞·ª£c (list, tuple, set, dict, v.v.) |
| `key`      | H√†m x√°c ƒë·ªãnh ti√™u ch√≠ so s√°nh (th∆∞·ªùng d√πng `lambda`)     |
| `reverse`  | `False` ‚Üí tƒÉng d·∫ßn (m·∫∑c ƒë·ªãnh), `True` ‚Üí gi·∫£m d·∫ßn         |

---

### üî• `lambda` trong `sorted()`
üîç Hi·ªÉu r√µ lambda trong sorted()

key=lambda x: <bi·ªÉu_th·ª©c>

x ƒë·∫°i di·ªán cho m·ªói ph·∫ßn t·ª≠ trong d·ªØ li·ªáu.

lambda tr√≠ch ra kh√≥a ƒë·ªÉ Python l·∫•y l√†m ti√™u ch√≠ s·∫Øp x·∫øp.

`lambda` gi√∫p ch·ªâ r√µ **s·∫Øp x·∫øp theo gi√° tr·ªã n√†o**.

#### 1) S·∫Øp x·∫øp **ƒë·ªëi v·ªõi dict.items()**

Khi duy·ªát m·ªôt dictionary b·∫±ng `.items()`, m·ªói ph·∫ßn t·ª≠ l√† m·ªôt tuple d·∫°ng `(key, value)`:

```python
(name, score)
```

Mu·ªën s·∫Øp x·∫øp theo **gi√° tr·ªã** (`score`), ta d√πng:

```python
sorted(scores.items(), key=lambda x: x[1])
```

* `x` l√† t·ª´ng ph·∫ßn t·ª≠ d·∫°ng `(name, score)`
* `x[1]` l√† ph·∫ßn **gi√° tr·ªã** d√πng ƒë·ªÉ so s√°nh

#### N·∫øu mu·ªën s·∫Øp x·∫øp **gi·∫£m d·∫ßn**, th√™m `reverse=True`:

```python
sorted(scores.items(), key=lambda x: x[1], reverse=True)
```
---

### T√≥m t·∫Øt

| M·ª•c ti√™u                          | C√¢u l·ªánh                                                   |
| --------------------------------- | ---------------------------------------------------------- |
| S·∫Øp x·∫øp **tƒÉng d·∫ßn theo gi√° tr·ªã** | `sorted(scores.items(), key=lambda x: x[1])`               |
| S·∫Øp x·∫øp **gi·∫£m d·∫ßn theo gi√° tr·ªã** | `sorted(scores.items(), key=lambda x: x[1], reverse=True)` |
| Chuy·ªÉn v·ªÅ `dict` sau s·∫Øp x·∫øp      | `dict(sorted(...))`                                        |
```


`````python
#  B√ÄI 1 ‚Äî Qu·∫£n l√Ω ƒëi·ªÉm h·ªçc sinh
# -------------------------------------------------------------
# Y√™u c·∫ßu:
# 1. Cho danh s√°ch ƒëi·ªÉm h·ªçc sinh d∆∞·ªõi d·∫°ng dict:
#    {"An": 8.5, "B√¨nh": 7.2, "Chi": 9.0, "D≈©ng": 6.8, "H√†": 8.0}
# 2. Vi·∫øt ch∆∞∆°ng tr√¨nh th·ª±c hi·ªán c√°c y√™u c·∫ßu sau:
#    - In danh s√°ch h·ªçc sinh v√† ƒëi·ªÉm c·ªßa h·ªç.
#    - T√¨m h·ªçc sinh c√≥ ƒëi·ªÉm cao nh·∫•t v√† th·∫•p nh·∫•t.
#    - T√≠nh ƒëi·ªÉm trung b√¨nh c·ªßa c·∫£ l·ªõp.
#    - T·∫°o m·ªôt dict m·ªõi l∆∞u x·∫øp lo·∫°i (>=8: "Gi·ªèi", 6.5‚Äì7.9: "Kh√°", <6.5: "Trung b√¨nh").
#    - S·∫Øp x·∫øp danh s√°ch h·ªçc sinh theo ƒëi·ªÉm gi·∫£m d·∫ßn.
# -------------------------------------------------------------
`````

`````python
danhsachSV={"An": 4.5, "B√¨nh": 7.2, "Chi": 9.0, "D≈©ng": 6.8, "H√†": 8.0}
print(f"{'t√™n':<20} {'ƒëi·ªÉm':<20}")
for ten,diem in danhsachSV.items():
    print(f"{ten:<20} {diem:<20}")
tenm=""
diemmax=0
for ten,diem in danhsachSV.items():
    if diemmax<diem:
        diemmax=diem
        tenm=ten
print("hoc sinh co diem cao nhat la: ",tenm)
for ten,diem in danhsachSV.items():
    if diemmax>diem:
        diemmax=diem
        tenm=ten
print("hoc sinh co diem thap nhat la: ",tenm)
tong=0
dem=0
for diem in danhsachSV.values():
    tong=tong+diem
    dem+=1
print("diem trung binh cua cac sinh vien la: ",tong/dem)
gioi={}
kha={}
tb={}
for ten,diem in danhsachSV.items():
    if diem>=8:
        gioi[ten]="gioi"

    elif diem>=6.5:
        kha[ten]="kha"

    else:
        tb[ten]="TB"


print("cac hojc sinh gioi",gioi)
print("cac hojc sinh kha",kha)
print("cac hojc sinh tb",tb)





`````
